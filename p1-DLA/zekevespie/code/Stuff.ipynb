```python
import numpy as np
import random

particles = 0
radius = 35 #edge radius
size = 101 #size of the grid
x0 = int((size-1)/2)
tree = np.zeros([size,size]) # n X n array means array center is [(n-1)/2,(n-1)/2] when n is odd
tree[x0,x0] = 1

```

```python
#Function to return the random point on the "edge"

def Spawn():
    rand1 = random.randrange(360)
    angle = rand1 * np.pi / 180
    x = radius * np.cos(angle) - x0 - 1
    y = radius * np.sin(angle) - x0 - 1
    print("x =", round(x), "| y =", round(y))
    return round(x), round(y)

i=1
while i<1000:
    x = 0
    y = 0
    x, y = Spawn()
#    print(x,y)
    tree[x, y] = 1
    i+=1
#    print(tree)

plt.matshow(tree)
```

```python
#Move the particle to the aggregate

print((x - x0)*(x - x0) + (y - x0)*(y - x0))
i=0
while i < 10000:
    if np.sqrt((x - x0)*(x - x0) + (y - x0)*(y - x0)) > 1.2*radius:
        break
    elif tree[x + 1, y] == 1 or tree[x - 1, y] == 1 or tree[x, y + 1] == 1 or tree[x, y - 1] == 1:
        particles += 1
        break
    else:
        print("##############################################################")
        print("##############################################################")
        print("##############################################################")
        print(tree)
        print("##############################################################")
        print("##############################################################")
        print("##############################################################")
        
        rand2 = random.randrange(4)
        if rand2 == 0:
            tree[x, y] = 0
            x+=1
            tree[x, y] = 1
            print("down")
        elif rand2 == 1:
            tree[x, y] = 0
            x-=1
            tree[x, y] = 1
            print("up")
        elif rand2 == 2:
            tree[x, y] = 0
            y+=1
            tree[x, y] = 1
            print("right")
        elif rand2 == 3:
            tree[x, y] = 0
            y-=1
            tree[x, y] = 1
            print("left")
        
        print(tree)
        i+=1

print((x - x0)*(x - x0) + (y - x0)*(y - x0))
```
